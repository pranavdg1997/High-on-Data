```python
import numpy as np
import pandas as pd
import twint
import nest_asyncio
nest_asyncio.apply()
```

```python
# Setting keywords just filter drug related tweets, we scrape
keywords = ["codeine","fentanyl","hydrocodone","oxycodone","Oxycontin","Percocet","Vicodin"]
df_scraped = pd.DataFrame()
for k in keywords:
    c = twint.Config()
    c.Store_csv = True
    
    #Setting time intervals to scrape from
    c.Since = "2019-9-01"
    c.Until = "2019-9-13"
    c.Location = True
    c.Output = "scraped_tweets_twint.csv"
    c.Search = k
    
    #limiting to 500 tweets for now
    c.Limit  = 500
    c.Lang = "en"
    c.Retries_count = 2
    twint.run.Search(c)
    curr_df = pd.read_csv("scraped_tweets_twint.csv")
    df_scraped = pd.concat((df_scraped,curr_df),axis=0).drop_duplicates()

# Collecting tweets irrespective of any topic just to make a mixed dataset
c = twint.Config()
c.Store_csv = True
c.Since = "2019-9-01"
c.Until = "2019-9-13"
c.Location = True
c.Output = "scraped_tweets_twint.csv"
c.Limit  = 500
c.Lang = "en"
c.Retries_count = 2
twint.run.Search(c)
curr_df = pd.read_csv("scraped_tweets_twint.csv")
df_scraped = pd.concat((df_scraped,curr_df),axis=0).drop_duplicates()
```
